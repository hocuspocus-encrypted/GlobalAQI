### <span style = "color:#0969DA">Introduction</span>

#### Air Pollution is contamination of the indoor or outdoor environment by any chemical, physical or biological agent that modifies the natural characteristics of the atmosphere. Household combustion devices, motor vehicles, industrial facilities and forest fires are common sources of air pollution. Pollutants of major public health concern include particulate matter, carbon monoxide, ozone, nitrogen dioxide and sulfur dioxide. Outdoor and indoor air pollution cause respiratory and other diseases and are important sources of morbidity and mortality. 

![image info](https://i.guim.co.uk/img/media/c20db72a9dcee4e16044e5733e235707411d35af/0_187_5640_3385/master/5640.jpg?width=465&dpr=1&s=none)

#### The dataset provides information about the following pollutants:

##### <span style = "color:orange"> Nitrogen Dioxide [NO2] : </span> 
> Nitrogen Dioxide is one of the several nitrogen oxides. It is introduced into the air by natural phenomena like entry from stratosphere or lighting. At the surface level, however, NO2 forms from cars, trucks and buses emissions, power plants and off-road equipment. Exposure over short periods can aggravate respiratory diseases, like asthma. Longer exposures may contribute to development of asthma and respiratory infections. People with asthma, children and the elderly are at greater risk for the health effects of NO2.

##### <span style = "color:orange"> Ozone [O3] : </span>
> The Ozone molecule is harmful for outdoor air quality (if outside the ozone layer). At surface level, ozone is created by chemical reactions between oxides of nitrogen and volatile organic compounds (VOC). Differently from the good ozone located in the upper atmosphere, ground level ozone can provoke several health problems like chest pain, coughing, throat irritation and airway inflammation. Furthermore, it can reduce lung function and worsen bronchitis, emphysema, and asthma. Ozone affects also vegetation and ecosystems. In particular, it damages sensitive vegetation during the growing season.

##### <span style = "color:orange"> Carbon Monoxide [CO] : </span>
> Carbon Monoxide is a colorless and odorless gas. Outdoor, it is emitted in the air above all by cars, trucks and other vehicles or machineries that burn fossil fuels. Such items like kerosene and gas space heaters, gas stoves also release CO affecting indoor air quality. Breathing air with a high concentration of CO reduces the amount of oxygen that can be transported in the blood stream to critical organs like the heart and brain. At very high levels, which are not likely to occur outdoor but which are possible in enclosed environments. CO can cause dizziness, confusion, unconsciousness and death.

##### <span style = "color:orange"> Particulate Matter [PM2.5] : </span>
> Atmospheric Particulate Matter, also known as atmospheric aerosol particles, are complex mixtures of small solid and liquid matter that get into the air. If inhaled they can cause serious heart and lungs problem. They have been classified as group 1 carcinogen by the International Agency for Research on Cancer (IARC). PM10 refers to those particles with a diameter of 10 micrometers or less. PM2.5 refers to those particles with a diameter of 2.5 micrometers or less.

#### <span style = "color:#0969DA">Importing the data</span>

```python
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Importing data from disk
aqi = pd.DataFrame(pd.read_csv('Assignment/global aqi/global air pollution dataset.csv'))
```

#### <span style = "color:#0969DA">Structure of the dataset</span>

```python
# Understanding the structure of the data
print(aqi.shape)
display(aqi)
```

#### <span style = "color:#0969DA">Data Cleaning</span>

```python
#Changing formatting
pd.set_option('display.precision', 1)
aqi.columns = ["Country", "City", "AQIValue", "AQICategory", "COValue", "COCategory", "OzoneValue", "OzoneCategory", "NO2Value", "NO2Category", "PMValue", "PMCategory"]
# Last 5 rows
aqi.tail()
```

#### <span style = "color:#0969DA">Exploratory Analyses</span>

```python
#Descriptive statistics(numerical values only)
exploratory = pd.DataFrame(aqi.describe())
print(exploratory)

print("Average Values : ", exploratory.loc["mean"])
print("Lowest Values : ", exploratory.loc["min"])
print("Highest Value : ", exploratory.loc["max"])
print("Median Values : ", exploratory.loc["50%"])
```

#### <span style = "color:#0969DA">Analysis</span>

```python
# Number of entries for each country
aqi['Country'].value_counts().sort_index().plot(kind='bar', color=['skyblue', 'lightcoral'], fontsize = 4)

plt.title("Number of entries per Country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation = 90, ha = 'right')
plt.gcf().set_size_inches(11,6)

plt.show()
```

#### <span style = "color:#0969DA">Exploratory plots</span>

```python
# Exploratory data plots
h = aqi.hist(bins = 'Country'.__sizeof__(), figsize=(15,15), grid = False)
```

#### <span style = "color:#0969DA">AQI Value in different countries</span>

```python
# Average AQI Value of different countries
sns.set(style = "whitegrid")

plt.figure(figsize=(6, 6))
sns.barplot(x='AQIValue', y='Country', data=aqi, palette='pastel')
plt.gcf().set_size_inches(10, 25)

plt.title("AQI Value in different countries")
```

#### <span style = "color:#0969DA">Grouped Bar Plot by Country for Various Pollutants</span>

```python
# Plot comparing values of various pollutants by country
group = aqi.groupby('Country')[['AQIValue', 'COValue', 'NO2Value', 'OzoneValue', 'PMValue']].mean()
print(group)

plt.rcParams['figure.figsize']=[50,10]
group.plot(kind = 'bar', colormap='tab20')

plt.title('Grouped Bar Plot by Country for Various Pollutants', fontsize = 50)
plt.xlabel('Country', fontsize = 25)
plt.ylabel('Value', fontsize = 25)

plt.legend(title='Pollutants', bbox_to_anchor=(1, 1), fontsize = 25)
plt.show()
```

#### <span style = "color:#0969DA">Scatter plots between AQI Value and various Pollutants</span>

```python
# Scatter plots to show relation between AQI Value and level of air pollutants
aqi.plot(x = "AQIValue", y = "COValue", kind = 'scatter', marker = '.', fontsize = 20, title = "AQI Value vs CO Value")
aqi.plot(x = "AQIValue", y = "NO2Value", kind = 'scatter', marker = '.', fontsize = 20, title = "AQI Value vs NO2 Value")
aqi.plot(x = "AQIValue", y = "OzoneValue", kind = 'scatter', marker = '.', fontsize = 20, title = "AQI Value vs Ozone Value")
aqi.plot(x = "AQIValue", y = "PMValue", kind = 'scatter', marker = '.', fontsize = 20, title = "AQI Value vs PM2.5 Value")
```

### <span style = "color:#0969DA">Conclusion</span>

Ultimately, the evidence strongly supports a clear correlation between the amount of air pollutants in the atmosphere and the Air Quality Index (AQI) values. Numerous studies and extensive air quality monitoring have consistently shown that as the concentration of pollutants such as particulate matter (PM), nitrogen dioxide (NO2), carbon monoxide (CO), and ozone (O3) increases, the AQI values also rise. This correlation underscores the direct impact of air pollutants on overall air quality.

High levels of air pollutants pose significant risks to human health, leading to respiratory and cardiovascular issues, among other adverse effects. The AQI serves as a crucial tool for communicating these risks to the public, providing a standardized and easily understandable measure of air quality. Monitoring and analyzing the correlation between air pollutants and AQI values contribute to our understanding of the environmental factors influencing air quality.

Efforts to mitigate air pollution and improve air quality must take into account this established correlation. Implementing effective pollution control measures, promoting sustainable practices, and raising awareness about the consequences of poor air quality are essential steps towards safeguarding public health and the environment. Recognizing and addressing the link between air pollutants and AQI values is imperative for developing informed policies and interventions to ensure cleaner, healthier air for present and future generations.
